# Architecture

```
class Net(nn.Module):
    def __int__(self, input_dim, hidden_dim, output_dim)
        super(Net,self).__init__()

        self.layers = nn.Sequential(nn.Linear(input_dim, hidden_dim),
                                     nn.ReLU(),
                                    nn.Linear(hidden_dim, hidden_dim),
                                     nn.ReLU(),
                                    nn.Linear(hidden_dim, hidden_dim),
                                    nn.ReLU(),
                                    nn.Linear(hidden_dim, hidden_dim),
                                    nn.ReLU(),
                                   nn.Linear(hidden_dim, output_dim),
                                   nn.Sigmoid())
        def forward(self, x):
            x = self.layers(x)
            return x
```
