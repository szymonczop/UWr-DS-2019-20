```python
def boyd_example_func(x, order=0):
    a=np.array([1,3])
    b=np.array([1,-3])
    c=np.array([-1,0])
    x=np.array(x)
    value = np.exp(a@x)+np.exp(b@x)+np.exp(c@x)
    if order==0:
        return value
    elif order==1:
        gradient = a.T*np.exp(a@x)+b.T*np.exp(b@x-0.1)+c.T*np.exp(c@x)
        return (value, gradient)
    elif order==2:
        gradient = a.T*np.exp(a@x)+b.T*np.exp(b@x)+c.T*np.exp(c@x)
        hessian = a.T*a*np.exp(a@x)+b.T*b*np.exp(b@x)+c.T*c*np.exp(c@x)
        return (value, gradient, hessian)
    else:
        raise ValueError("The argument order should be 0, 1 or 2")
```

```python
def quadratic( H, b, x, order=0 ):
    """ 
    Quadratic Objective
    H:          the Hessian matrix
    b:          the vector of linear coefficients
    x:          the current iterate
    order:      the order of the oracle. For example, order=1 returns the value of the function and its gradient while order=2 will also return the hessian
    """
    #H = np.asmatrix(H)
    b = np.array(b)
    x = np.array(x)

    value = 0.5 * x.T @ H @ x + b.T @ x # to się zawsze zgadza 

    if order == 0:
        return value
    elif order == 1:
        gradient = H @ x + b
        return (value, gradient)
    elif order == 2:
        gradient = H @ x + b
        hessian = H
        return (value, gradient, hessian)
    else:
        raise ValueError("The argument \"order\" should be 0, 1 or 2")
```

### Backtracking for my function 
$ f(x) = x_1^4 + x_2^4 + x_1^2x_2 + x_3^2$

```python
def back_track_multi(function,gradient_function,alpha,beta,start_x):
    t =1
    direction =np.sign(-gradient_function(start_x)) #np.sign(- grad_fun_multi(start_x))
    while function(start_x + t*direction)> function(start_x)+ alpha*t*gradient_function(start_x).dot(direction):
        t = beta*t
        #print(f" Wartość funkcji {fun_multi(start_x + t*direction)} t-value = {t},dir = {direction}")
    return t 
```

```python
def back_track_multi_boyd(function,alpha,beta,start_x):
    t =1
    direction =np.sign(-function(start_x,order = 1)) #np.sign(- grad_fun_multi(start_x))
    while function(start_x + t*direction,order = 0)> function(start_x,order = 0 )+ alpha*t*function(start_x,order = 1).dot(direction):
        t = beta*t
        #print(f" Wartość funkcji {fun_multi(start_x + t*direction)} t-value = {t},dir = {direction}")
    return t 
```

 ### Boyd

```python
def gradient_descent(gradient_function, start_x, eps = 1e-5, maximum_iterations=1000): 

    steps = [start_x]
    i = 0
    while True:
        i +=1
        direc = np.sign(-gradient_function(start_x,order = 1))
        grad_length = gradient_function(start_x,order = 1) @ gradient_function(start_x,order = 1)
        t = back_track_multi_boyd(my_boyd_example_func,0.1,0.5,start_x)
        start_x = start_x + t*direc
        steps.append(start_x)
        if(grad_length < eps or i > maximum_iterations ):
            break
        #print(f"my points {i},I'm in point {start_x}")
    return start_x, i ,steps
```
