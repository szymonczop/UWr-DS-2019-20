```python
import time
import numpy as np
import matplotlib.pyplot as plt 

def rosenbrock_v(x):
    """Returns the value of Rosenbrock's function at x"""
    return (1 - x[0])**2 + 100*(x[1]- x[0]**2)**2



def rosenbrock_hessian(x):
    #TODO: compute the value, gradient and Hessian of Rosenbrock's function'
        val = rosenbrock_v(x)
        dVdX= np.array([2*(200*x[0]**3 - 200*x[0]*x[1] + x[0]-1),200*(x[1]-x[0]**2)])
        H = np.array([[1200*x[0]**2-400*x[1] + 2,-400*x[0]],
                      [-400*x[0],200]])
        return [val,dVdX, H]

def back_track_multi(function,alpha,beta,start_x):
    t =1
    direction =np.sign(-function(start_x)[1]) #np.sign(- grad_fun_multi(start_x))
    while function(start_x + t*direction)[0]> function(start_x)[0]+ alpha*t*function(start_x)[1].dot(direction):
        t = beta*t
        #print(f" Wartość funkcji {fun_multi(start_x + t*direction)} t-value = {t},dir = {direction}")
    return t 
```

##  backtracking

```python
def Newton_no(f,initial_x, eps =1e-2, maximum_iterations=10000):
    if eps <= 0:
        raise ValueError("Epsilon must be positive")
    
    # initialization
    values = []
    runtimes = []
    #xs = []
    start_time = time.time()
    i = 0  
    start_x = initial_x
    
    while True:
        i += 1
        print(f"jestem w {i} iteracji i punkcie {start_x} ")
        grad = f(start_x)[1]
        H = f(start_x)[2]
        
        values.append(start_x)
        runtimes.append( time.time() - start_time )
        
        direc_len = -(np.linalg.inv(H) @ grad)
        
        newton_decrement = grad @ np.linalg.inv(H) @ grad.T
        
        if (newton_decrement**2)/2 < eps:
            break
        
        t = back_track_multi(f,0.25,0.5,start_x)
        print(f" value of my step t = {t}")
        
        start_x = start_x + t*direc_len
        
        if i > maximum_iterations or t <eps:
            break
    Theta = start_x
        
    return Theta, values, runtimes, i
```

```python
x_start = [0,2]
#Newton_no(rosenbrock_hessian, x_start)
        
        
Xopt, Xhist,step_time,fun_evals = Newton_no(rosenbrock_hessian, x_start)


Xhist_ = np.array([[x[0], x[1]] for x in Xhist])


MX,MY = np.meshgrid(np.linspace(-2,2,100), np.linspace(-2,2,100))
Z = np.array([MX,MY]).reshape(2,-1)
VR = rosenbrock_v(Z)
plt.contour(MX,MY,VR.reshape(MX.shape), 100)
plt.plot(Xhist_[:,0], Xhist_[:,1], '*-k')
```

```python
def exact_line_search(f,x_start,direc_len,alpha = 1e-2):
    t = 1
    direc_len = direc_len*alpha
    x_start = x_start + direc_len
    x_step_forward = x_start + direc_len
    while f(x_step_forward)[0]<f(x_start)[0]:
        x_start = x_step_forward
        x_step_forward = x_start + direc_len
        t += 1
    return alpha*t
```

# QUADRATIC

```python
H = np.array([[1,-3],
              [8,32]])
    
b = np.array([-1,3])


def quadratic_v(x):
    value = 1/2*x[0]**2 + 5/2*x[0]*x[1] + 16*x[1]**2 - x[0] + 3*x[1]
    return value



def quadratic(x):
    """ 
    Quadratic Objective
    H:          the Hessian matrix
    b:          the vector of linear coefficients
    x:          the current iterate
    order:      the order of the oracle. For example, order=1 returns the value of the function and its gradient while order=2 will also return the hessian
    """
    x = np.array(x)
    value = 0.5 * x.T @ H @ x + b.T @ x # to się zawsze zgadza 
    gradient = H @ x + b
    hessian = H
    return (value, gradient, hessian)
```

# CG 

```python
def cg(f,initial_x,eps = 1e-2,max_iteration = 1000,stopper = 1e-2):
    if eps <= 0:
        raise ValueError("Epsilon must be positive")
    
    values = []
    runtimes  = []
    
    start_time = time.time()
    i = 0
    start_x = initial_x
    p_0 = -f(start_x)[1]
    while True:
        
        values.append(start_x)
        runtimes.append( time.time() - start_time )
        
        t = back_track_multi(f,0.25,0.5,start_x)
        x_up = start_x + t*p_0*stopper
        
        print(f"Im in {i} iteration with step value {t} x-coordination {start_x}")
        
        grad_up = f(x_up)[1]
        beta = (grad_up @ grad_up) / (p_0 @ p_0)
        
        print(f"moja beta {beta}")
        print(f"moj stop condition {p_0@p_0}")
        
        if i % 2 == 0 : 
            beta = 0 
            
        
        p_up = -grad_up + beta*p_0
        
        p_0 = p_up
        start_x = x_up
        
        i += 1
        
        
        if (p_0 @ p_0 < eps or i > max_iteration or t < eps):
            break 
    
    Theta = start_x
    
    return Theta, values, runtimes ,i
         
```

```python
def cgpr(f,initial_x,eps = 1e-5,max_iteration = 1000,stopper = 1e-2):
    if eps <= 0:
        raise ValueError("Epsilon must be positive")
    
    values = []
    runtimes  = []
    
    start_time = time.time()
    i = 0
    start_x = initial_x
    p_0 = -f(start_x)[1]
    while True:
        
        values.append(start_x)
        runtimes.append( time.time() - start_time )
        
        t = back_track_multi(f,0.25,0.5,start_x)
        x_up = start_x + t*p_0*stopper
        
        print(f"Im in {i} iteration with step value {t} x-coordination {start_x}")
        
        grad_up = f(x_up)[1]
        beta = (grad_up @ (grad_up - p_0)) / (p_0 @ p_0)
        
        print(f"moja beta {beta}")
        print(f"moj stop condition {p_0@p_0}")
        
        if i % 2 == 0 : 
            beta = 0 
            
        
        p_up = -grad_up + beta*p_0
        
        p_0 = p_up
        start_x = x_up
        
        i += 1
        
        
        if (p_0 @ p_0 < eps or i > max_iteration or t < eps):
            break 
    
    Theta = start_x
    
    return Theta, values, runtimes ,i
```

```python
def powell(x):
    val = (x[0] + 10*x[1])**2 + 5*(x[2] - x[3])**2 + (x[1] - 2*x[2])**4 + 10*(x[0] - x[3])**4
    grad = np.array([2*(20*((x[0] - x[3])**3) + x[0] + 10*x[1]), 4*(5*(x[0]+10*x[1])+(x[1]-2*x[2])**3),10*(x[2]-x[3])-8*(x[1]-2*x[2])**3,10*(-4*(x[0]-x[3])**3+x[3]-x[2])])
    return val, grad
```

```python
def cg2_powell(f,initial_x,eps = 1e-5,max_iteration = 1000,stopper = 1e-2):
    if eps <= 0:
        raise ValueError("Epsilon must be positive")
    
    values = []
    runtimes  = []
    
    start_time = time.time()
    i = 0
    start_x = initial_x
    p_0 = -f(start_x)[1]
    while True:
        
        values.append(start_x)
        runtimes.append( time.time() - start_time )
        
        t = exact_line_search(f,start_x,p_0,alpha = 1e-3) # to jest zmienione w tej funkcji
        x_up = start_x + t*p_0
        
        print(f"Im in {i} iteration with step value {t} x-coordination {start_x}")
        
        grad_up = f(x_up)[1]
        beta = (grad_up @ grad_up) / (p_0 @ p_0)
        
        print(f"moja beta {beta}")
        print(f"moj stop condition {p_0@p_0}")
        
        if i % 4 == 0 : # important changes
            beta = 0 
            
        
        p_up = -grad_up + beta*p_0
        
        p_0 = p_up
        start_x = x_up
        
        i += 1
        
        
        if (p_0 @ p_0 < eps or i > max_iteration or t < eps):
            break 
    
    Theta = start_x
    
    return Theta, values, runtimes ,i
         
```
