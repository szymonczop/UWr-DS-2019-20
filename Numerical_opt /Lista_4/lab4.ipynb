```python
def gradient_descent(x, fun1, alpha=1.0, eps=0, iters=300):
    x=np.array(x,dtype=float)
    i=0
    while iters>0:
        iters-=1
        i+=1
        direction=np.array(fun1(x))
        #print(direction,direction@direction)
        #print(alpha,x)
        x-=alpha*direction
        step_length=direction@direction
        if(  step_length < eps or step_length==float('inf')  ):
            break
    return x,i
```

# problem 4.2

```python
def boyd_example_func(x, order=0):
    a=np.array([1,3])
    b=np.array([1,-3])
    c=np.array([-1,0])
    x=np.array(x)
    #a,b,c,x=np.array(a),np.array(b),np.array(c),np.array(x)
    value = np.exp(a@x)+np.exp(b@x)+np.exp(c@x)
    if order==0:
        return value
    elif order==1:
        gradient = a.T*np.exp(a@x)+b.T*np.exp(b@x-0.1)+c.T*np.exp(c@x)
        return (value, gradient)
    elif order==2:
        gradient = a.T*np.exp(a@x)+b.T*np.exp(b@x)+c.T*np.exp(c@x)
        hessian = a.T*a*np.exp(a@x)+b.T*b*np.exp(b@x)+c.T*c*np.exp(c@x)
        return (value, gradient, hessian)
    else:
        raise ValueError("The argument order should be 0, 1 or 2")
```

```python
def fun1(x):
    return boyd_example_func(x,order=1)[1]
def fun(x):
    return boyd_example_func(x)
fun([1,2])
```
